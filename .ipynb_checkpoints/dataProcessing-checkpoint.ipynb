### Data Cleaning and Exploratory Analysis  -- 2012 Commodity Flow Survey (CFS)

```python
import pandas as pd
import matplotlib
import numpy as np
import os

import matplotlib.pyplot as plt
matplotlib.style.use('classic')
%matplotlib inline

from sklearn.preprocessing import scale
```

```python
filePath = "data/cfs_2012_pumf_csv.csv"
cfs2012 = pd.read_csv(filePath, sep=',', index_col=0)
print len(cfs2012.index)
cfs2012.head()
```

Analysis will focus on truck trips only, so filter the entries accordingly. According to the CFS data dictionary:

Mode Code 4 - For-Hire truck

Mode Code 5 - Private truck

```python
cfs = cfs2012.loc[(cfs2012["MODE"] == 4) | (cfs2012["MODE"] == 5)]
cfs = cfs.loc[(cfs["SHIPMT_DIST_ROUTED"] != 0) | (cfs["SHIPMT_DIST_ROUTED"] != 99999)] #Drop trips with missing or invalid data
print len(cfs.index)
cfs.head(20) 
```

Judging from the length of dataframes, it appears that the majority of shipments recorded are in fact delivered via truck. Next look at a histogram of the routed distance between origin and destination points.

```python
hist = cfs["SHIPMT_DIST_ROUTED"].plot.hist(bins=50, figsize=(20,10), normed=True, cumulative=False, 
                                           title="Histogram of Truck Delivery Distances", 
                                           xticks=range(0,5000,200), grid=True)
hist.set_xlabel("Routed Shipment Distance (mi)")
```

```python
cumHist = cfs["SHIPMT_DIST_ROUTED"].plot.hist(bins=50, figsize=(20,10), normed=True, cumulative=True, 
                                              title="Cumulative Histogram of Truck Delivery Distances", histtype='step',
                                              xticks=range(0,5000,200), grid=True)
cumHist.set_xlabel("Routed Shipment Distance (mi)")
```

Interestingly, most of the trips are within the purported 200-300 mile range of upcoming electric semi-trucks: around 75% of deliveries in the survey have routed distances of less than 300 miles. Keep in mind that this is only representative of the sample. The CFS provides a weighting factor for each entry, which is used to estimate the number of actual deliveries each sample represents. To truly tell how many trips can be completed with a 200-300 mile range the weighting factor need be applied.

```python
def trip_pct_report(cfs, capDistance):
    totalTruckMiles = np.sum([dist*wght for dist,wght in zip(cfs["SHIPMT_DIST_ROUTED"], cfs["WGT_FACTOR"])])
    totalTruckTrips = np.sum(cfs["WGT_FACTOR"])
    cfsChoped = cfs.loc[cfs["SHIPMT_DIST_ROUTED"] <= capDistance] #Only include trips under capDistance
    shortTruckMiles = np.sum([dist*wght for dist,wght in zip(cfsChoped["SHIPMT_DIST_ROUTED"], cfsChoped["WGT_FACTOR"])])
    shortTruckTrips = np.sum(cfsChoped["WGT_FACTOR"])
    shortTripMilesPct = (shortTruckMiles / totalTruckMiles)*100.0
    shortTripPct = (shortTruckTrips / totalTruckTrips)*100.0
    print "Total truck delivery miles: %d" % totalTruckMiles
    print "Total number of truck deliveries: %d" % totalTruckTrips
    print "Total truck delivery miles for trips %s miles or less: %d" % (capDistance, shortTruckMiles)
    print "Total number of truck deliveries for trips %s miles or less: %d" % (capDistance, shortTruckTrips)
    print "Percentage of total truck delivery miles for trips %s miles or less: %.2f percent" % (capDistance, shortTripMilesPct)
    print "Percentage of total truck trips %s miles or less: %.2f percent" % (capDistance, shortTripPct)
```

```python
trip_pct_report(cfs, 300)
```

Now it can be seen that while the majority of the truck trips completed were less than 300 miles, these trips only represent a small portion of the total miles driven by trucks deliverying commodities; however, a reduction in 16 pct of truck delivery emissions (assuming equivalent emissions/mile ratio for short and long trips) still represents a significant benefit. 

Lets take a closer look at the distance distribution of trips 300 miles or less.

```python
cfsShort = cfs.loc[cfs["SHIPMT_DIST_ROUTED"] <= 300.0] #Only include trips under capDistance
hist = cfsShort["SHIPMT_DIST_ROUTED"].plot.hist(bins=50, figsize=(20,10), normed=True, cumulative=False, 
                                                title="Histogram of Truck Delivery Distances for Trips 300 Miles or Less", 
                                                xticks=range(0,300,10), grid=True)
hist.set_xlabel("Routed Shipment Distance (mi)")
```

```python
cumHist = cfsShort["SHIPMT_DIST_ROUTED"].plot.hist(bins=50, figsize=(20,10), normed=True, cumulative=True, 
                                        title="Cumulative Histogram of Truck Delivery Distances for Trips 300 Miles or Less", 
                                        histtype='step', xticks=range(0,300,10), grid=True)
cumHist.set_xlabel("Routed Shipment Distance (mi)")
```

When the distance distribution of shorter trips, in this case trips under 300 miles, are inspected, it can be seen that a significant majority of these trips are under 100 miles. Just under 80 percent of the delivery trips are under 100 miles in routed distance. Again, note this particular visualization is only indicative of the sample set, without considering the weighting factors. 

```python
trip_pct_report(cfsShort, 100)
```

A similar situation occurs for deliveries with routed distances less than or equal to 300 miles as occured when looking at the uncapped data set: most trips are short, but these trips do not represent a majority of the total miles driven by trucks delivering commodities. However, a much larger portion of the total miles is represented by shorter trips, which in this context are considered 100 miles or less in routed distance.

Let's take a look at the shipment weight distributions for delivering with routed distances 300 miles or less.

```python
tons = cfsShort["SHIPMT_WGHT"] / 2000.0 #Convert weight to tons
hist = tons.plot.hist(bins=50, figsize=(20,10), normed=False, cumulative=False, 
                      title="Histogram of Truck Shipment Weights for Trips 300 Miles or Less", 
                      grid=True,
                      logy=True) #Had to use a log scale to account for outliers
hist.set_xlabel("Shipment Weight (tons)")
```

Clearly the large spread of shipment weights, including outliers of very heavy shipments, are making histogram visualization difficult. Recently Cummins has released towing capabilities of its upcoming eletric semi: 44,000 lbs or 22 tons. Lets shrink the dataset to only include shipments of 30 tons or less and redo the histogram of shipment weights.

```python
tonsChop = tons[tons <= 30.0]
hist = tonsChop.plot.hist(bins=50, figsize=(20,10), normed=False, cumulative=False, 
                      title="Histogram of Truck Shipment Weights for Trips 300 Miles or Less", 
                      xticks=range(0,30,1), grid=True,
                      logy=False)
hist.set_xlabel("Shipment Weight (tons)")
```

It can be seen that the vast majority of shipments are under two tons. Lets dive into the data again and look at the histogram of for shipment weights of two tons or less.

```python
tonsChop = tons[tons <= 2.0]
hist = tonsChop.plot.hist(bins=50, figsize=(20,10), normed=False, cumulative=False, 
                      title="Histogram of Truck Shipment Weights (2 tons or less) for Trips 300 Miles or Less", 
                      xticks=[i/10.0 for i in range(21)], grid=True, logy=False)
hist.set_xlabel("Shipment Weight (tons)")
```

```python
cumHist = tonsChop.plot.hist(bins=50, figsize=(20,10), normed=True, cumulative=True, 
                      title="Cumulative Histogram of Truck Shipment Weights (2 tons or less) for Trips 300 Miles or Less", 
                      xticks=[i/10.0 for i in range(21)], grid=True, logy=False, histtype='step')
cumHist.set_xlabel("Shipment Weight (tons)")
```

A similar picture is seen in this histogram. Most shipments are light, with the vast majority (about 80%) being under 0.5 tons. 

```python
#Check if deliveries of 300 miles or less most often occur to urban areas or not. 
#Look at DEST_MA & ORIG_MA, if the value is not 99999 then assume the destination area is urban.
```

Typically stop-and-go driving found within urban areas provides worse fuel economy than highway driving. Lets look at the distribution of metro area deliveries compared to non-metro areas. Fortunately the CFS dataset provides a metro-area code "DEST_MA" and "ORIG_MA", for these features 99999 indicates an area outside of the large urban centers of each state. The data will be segregated accordingly, where "urban" shipments will considered those where both the origin and destination areas are metropolitan. 

```python
def urban_check(orig, dest):
    if orig != 99999 and dest != 99999:
        return "METRO"
    else:
        return "NON-METRO"
        
cfsShortMA = cfsShort.copy()
cfsShortMA["AREA_TYPE"] = cfsShortMA.apply(lambda row: urban_check(row["ORIG_MA"], row["DEST_MA"]), axis=1)
```

```python
MAGroup = cfsShortMA.groupby(["AREA_TYPE"])
MAGroupSize = MAGroup.size()
MAGroupSize.plot.bar(figsize=(20,10), title="Shipment Area Comparison for Truck Shipments 300 Miles or Less")
```

Most shipments with routed distances of 300 miles or less begin and end within metropolitan areas; although, the fraction is near half. 

### Conclusion

From the data, there seems to be an argument to be made for focusing on short light shipments with electric trucks with a priority being the number of cheaper electric trucks put into service rather than prioritizing more expensive but more capable (in terms of hauling weight and distance capabilities) trucks. Such an approach would have electric trucks performing delivering of light goods over short distances, which account for the majority of truck shipments in the US. However, this approach would not provide optimal benefit in terms of total truck delivery miles driven. A similar line of thinking supports deployment of delivery drones for short lightweight shipments, an approach being actively discussed by Amazon.

Assuming that very short trips provide worse fuel efficiency performance when compared to longer hauls, along the lines of city driving miles versus highway driving miles, electric trucks could provide great benefit in this area. Also short deliveries are likely to occur in urban centers, for instance from distribution centers to homes and businesses, emissions reduction in these areas could represent a significant public health benefit in these population dense zones.
